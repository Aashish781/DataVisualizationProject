# Data Visualization

#### Importing Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Reading dataset
df_png = pd.read_csv("PNG_Statistics_2020-21_Table5.22.csv")
```

```python
# Dataset info.
df_png.info()
```

```python
# Display all columns and rows
pd.set_option('display.max_columns', None)
df_png.head(41)
```

```python
# Stats related to dataset
df_png.iloc[:36].describe()
```

## Bar Plot

```python
# For year 2020-21
plt.figure(figsize=(20, 10))
plt.bar(df_png["State/UT"].iloc[:36], df_png["2020-21 (P)"].iloc[:36])
plt.title("State-wise Consumption of Total Petroleum Products for FY:2020-21", fontsize=20)
plt.ylabel(" Consumption (TMT) ", fontsize=18)
plt.xlabel("States",fontsize=18)
plt.xticks(rotation = 60)
plt.show()
```

```python
# For year 2019-20
plt.figure(figsize=(20, 10))
plt.bar(df_png["State/UT"].iloc[:36], df_png["2019-20"].iloc[:36], color="orange")
plt.title("State-wise Consumption of Total Petroleum Products for FY:2019-20")
plt.ylabel(" Consumption (TMT) ")
plt.xlabel("States")
plt.xticks(rotation = 60)
plt.show()
```

## Box Plot

```python
col = list(df_png.columns)
col = col[1:8]

yearwiseCons = []
for year in col:
    df_year = df_png[year].iloc[:36].to_numpy()
    yearwiseCons.append(df_year)
```

```python
# Box plot
plt.figure(figsize=(10, 8))
plt.boxplot(yearwiseCons,labels=col, patch_artist=True)
plt.title("Year wise Consumption of Total Petroleum Products")
plt.ylabel(" Consumption (TMT) ")
plt.xlabel(" Year")
plt.show()
```

## Scatter Plot

```python
bihar = list(df_png.loc[df_png['State/UT'] == "Bihar"].iloc[0])[1:8]
delhi = list(df_png.loc[df_png['State/UT'] == "Delhi"].iloc[0])[1:8]
kerala = list(df_png.loc[df_png['State/UT'] == "Kerala"].iloc[0])[1:8]
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(col,bihar,label="Bihar")
plt.scatter(col,delhi,label="Delhi")
plt.scatter(col,kerala,label="Kerala")
plt.title("Consumption of Total Petroleum Products")
plt.ylabel(" Consumption (TMT) ")
plt.xlabel(" Year")
# plt.xticks(rotation = 7)
plt.legend()
plt.show()
```
